```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 定義數據
groups = ["Control Group", "Experimental Group"]
means = [2.6845, 3.3166]
std_devs = [0.75698, 0.98436]

# 設置顏色（冷色系 & 暖色系）
colors = ["#6baed6", "#fc8d62"]  # 藍色（冷）& 橘色（暖）
error_colors = ["#2171b5", "#e6550d"]  # 深藍 & 深橘，用於誤差線
marker_colors = ["#08519c", "#a63603"]  # 更深的冷暖色


# 設置全局字體
rcParams['font.family'] = 'Times New Roman'  # 你可以改成 "Times New Roman", "SimHei"（黑體，適用中文）等
rcParams['font.size'] = 12  # 設定整體字體大小

# 建立圖表
plt.figure(figsize=(6, 4))

# 使用條形圖背景（不同顏色）
plt.bar(groups, means, color=colors, alpha=0.7, width=0.6, label="Mean") 

# 添加誤差線（不同顏色）
for i in range(len(groups)):
    plt.errorbar(groups[i], means[i], yerr=std_devs[i], fmt='o', capsize=8, capthick=2, markersize=10,
                 markerfacecolor=marker_colors[i], markeredgecolor="white", ecolor=error_colors[i], elinewidth=2)

# 設置標題與標籤
plt.xlabel("Group", fontsize=12)
plt.ylabel("Mean Score", fontsize=12)
plt.title("Descriptive Statistics of STEM Self-Efficacy Across Different Group", fontsize=14, fontweight='bold')

# 設置 X 軸標籤 & 網格線
plt.xticks(fontsize=11)
plt.grid(axis='y', linestyle="--", alpha=0.5)

# 顯示圖例
plt.legend(["CG", "EG"], frameon=False, fontsize=11)

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 定義數據
groups = ["Control Group", "Experimental Group"]
learning_units = ["GOES Solar", "Mercury Retrograde", "TESS Transit"]

# 取出計算出的均值與標準差
means_goes = [2.4039, 3.0037]  # GOES Solar Data
std_goes = [0.6622, 0.6858]

means_mercury = [2.6360, 2.5615]  # Mercury Retrograde
std_mercury = [0.2303, 0.4607]

means_tess = [2.6395, 3.5392]  # TESS Transit
std_tess = [0.2260, 0.8551]

# 設置顏色（冷色系 & 暖色系）
colors = ["#6baed6", "#fc8d62"]  # 藍色（冷）& 橘色（暖色）
error_colors = ["#2171b5", "#e6550d"]  # 深藍 & 深橘，用於誤差線
marker_colors = ["#08519c", "#a63603"]  # 更深的冷暖色

# 設置全局字體
rcParams['font.family'] = 'Times New Roman'
rcParams['font.size'] = 12

# 建立三個子圖
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# 定義圖表繪製函數
def plot_bar(ax, means, stds, title):
    ax.bar(groups, means, color=colors, alpha=0.7, width=0.6)
    for i in range(len(groups)):
        ax.errorbar(groups[i], means[i], yerr=stds[i], fmt='o', capsize=8, capthick=2, markersize=10,
                    markerfacecolor=marker_colors[i], markeredgecolor="white", ecolor=error_colors[i], elinewidth=2)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel("Group", fontsize=12)
    ax.grid(axis='y', linestyle="--", alpha=0.5)

# 繪製三個圖表
plot_bar(axes[0], means_goes, std_goes, "GOES Solar Data Analysis")
plot_bar(axes[1], means_mercury, std_mercury, "Mercury Retrograde Analysis")
plot_bar(axes[2], means_tess, std_tess, "TESS Transit Analysis")

# 設置 Y 軸標籤
axes[0].set_ylabel("Mean Score", fontsize=12)

# 顯示圖表
plt.tight_layout()
plt.show()
```
