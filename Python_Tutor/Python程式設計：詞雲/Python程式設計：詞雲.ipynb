```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

with open("/content/drive/MyDrive/python tutor/詞雲素材/Eenie Meenie歌詞.txt") as fp:  # 英文字的文字檔
    text = fp.read()              # 讀取檔案

wd = WordCloud().generate(text)   # 由txt文字產生WordCloud物件
imageCloud = wd.to_image()       # 由WordCloud物件建立詞雲影像檔

# 使用 Matplotlib 顯示圖片
plt.imshow(imageCloud, interpolation='bilinear')
plt.axis('off') # 關閉座標軸
plt.show() # 顯示圖片
```
